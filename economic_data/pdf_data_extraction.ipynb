## Extract Data From PDF using PDF Plumber library

### Imports

```python
import pdfplumber
import pandas as pd
```

### extract bordered table

```python
pdf_path = r'C:\Users\Wissem\Downloads\Hardship-Index-Fact-Sheet-2017-ACS-Final.pdf'
```

```python
def extract_tables_from_pdf(pdf_file):
    with pdfplumber.open(pdf_file) as pdf:
        tables = []
        for page in pdf.pages:
            tables.extend(page.extract_tables())
    return tables
```

```python
extracted_tables = extract_tables_from_pdf(pdf_path)
```

```python
real_tables = []
real_tables.append(extracted_tables[1])
real_tables.append(extracted_tables[2])
tables = []
```

```python
table = real_tables[1]
for i in range(2,len(table)):
    L = table[i][-2].split()
    table[i][-2] = L[0]
    table[i][-1] = L[1] + table[i][-1]

        
```

```python

for table in real_tables:
    table_data = {}
    table_data["columns"] = table[1]
    table_data["data"] = []
    for i in range(2,len(table)):
        table_data["data"].append(table[i])
    tables.append(table_data)
```

### Extract Borderless Table

```python
with pdfplumber.open(pdf_path) as pdf:
    borderless_table = pdf.pages[4].extract_text()
    borderless_table = borderless_table.split("\n")
```

```python
borderless_table.pop(0)
borderless_table.pop(0)
borderless_table.pop(0)
borderless_table.pop(0)
borderless_table.pop(-1)
```

```python
borderless_table
```

```python
table = []
for str_line in borderless_table:
    line = str_line.split()
    for index in range(1,len(line)):
        if line[index].isalpha():
            line[0] = line[0] + " " + line[index]
    table.append(line)

for row in table:    
    for index, item in enumerate(row):
        if index != 0 and len(item.split()) == 1 and item.isalpha():
            row.remove(item)
        
    
     
```

```python
table_3 = {}
table_3["columns"] = []
table_3["data"] = table
tables.append(table_3)
```
