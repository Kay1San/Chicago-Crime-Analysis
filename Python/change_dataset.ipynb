```python
import pandas as pd
```

```python
df = pd.read_csv("Kay_Dataset.csv")
df.head()
```

```python
# Assuming df is your DataFrame with a 'Date' column
# Convert the 'Date' column to datetime objects with an explicit format
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%dT%H:%M:%S.%f')

# Define the function to categorize the time of day
def categorize_time_of_day(dt):
    hour = dt.hour
    if 5 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 17:
        return 'Afternoon'
    elif 17 <= hour < 21:
        return 'Evening'
    else:
        return 'Night'

# Apply the function to categorize times of day
df['TimeOfDay'] = df['Date'].apply(categorize_time_of_day)

# If you want to replace the 'Date' column with these categories
df['Date'] = df['TimeOfDay']

# Drop the 'TimeOfDay' column if it's no longer needed
df.drop('TimeOfDay', axis=1, inplace=True)

# Save the updated DataFrame to a CSV, overwriting the original file
df.to_csv('Kay_Dataset.csv', index=False)
```
